```python
# Shell sort
import random
import time
import tracemalloc
key = {
    'a': 1, 'b': 2, 'c':3,'d':4,'e':5,'f':6,'g':7,'h':8,'i':9,'j':10, 'k':11, 'l':12, 'm':13, 'n':14,'o':15, 'p':16, 'q':17, 'r': 18, 's':19, 't':20, 'u':21, 'v':22,'w':23,'x':24,'y':25,'z':26
    }
def shellSort(array, n):
            count = 0
            interval = n // 2
            while interval > 0:
                count = count + 1
                for i in range(interval, n):
                    temp= array[i]
                    j = i
                    while j >=  interval and array[j - interval] > temp:
                        count = count + 1
                        array[j] = array[j - interval]
                        j -= interval
                    array[j] = temp
                interval //= 2
            print('count: ',count)
data = []
y = 1
ask = input('Do you want to test int or str or float?')
ask2 = 2**10  #list length
while y <= ask2 and ask == 'str':
    d=random.choice(list(key))
    data.append(d)
    y = y + 1
while y <= ask2 and ask == 'int':
    d=random.randrange(-1000000,1000000)
    data.append(d)
    y = y + 1

size = len(data)
print(data)
tracemalloc.start()
start= time.time()
x1, y1 = tracemalloc.get_traced_memory()
shellSort(data, size)
x2, y2 = tracemalloc.get_traced_memory()
end=time.time()
tracemalloc.stop()
print('Sorted Array in Ascending Order Or In Alphabetecal Order')
print(data)
print('time:',end-start)
print(y2 - y1)



```
